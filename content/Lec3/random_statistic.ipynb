## 4. random Library

The `random` library is used for random numbers, picking random items, and shuffling lists.

Common functions:
- `random.randint(a, b)` → random integer from a to b
- `random.choice(seq)` → pick one random element from a list
- `random.shuffle(list)` → shuffle the list order

```python
# Example 1: random integers
import random

for i in range(5):
    num = random.randint(1, 6)
    print("Dice roll:", num)
```

```python
# Example 2: random choice
import random

colors = ["red", "blue", "green", "yellow"]
pick = random.choice(colors)
print("Picked color:", pick)
```

```python
# Example 3: shuffle a list
import random

cards = ["A", "K", "Q", "J", "10"]
random.shuffle(cards)
print("Shuffled cards:", cards)
```

## 5. statistics Library

The `statistics` library helps you calculate things like average, median, and standard deviation.

```python
# Example: basic average
import statistics

scores = [80, 90, 70, 100, 85]
avg = statistics.mean(scores)
print("Average score:", avg)
```

```python
# Example: median and standard deviation
import statistics

data = [2, 4, 4, 4, 5, 5, 7, 9]
print("Median:", statistics.median(data))
print("Standard deviation:", statistics.stdev(data))
```
