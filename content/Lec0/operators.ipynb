# Operators
### 1. Arithmetic Operators


| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `+` | Addition | `5 + 3` | `8` |
| `-` | Subtraction | `10 - 4` | `6` |
| `*` | Multiplication | `7 * 2` | `14` |
| `/` | Division (float) | `8 / 2` | `4.0` |
| `//` | Floor Division | `7 // 2` | `3` |
| `%` | Modulus (remainder) | `7 % 2` | `1` |
| `**` | Exponentiation | `2 ** 3` | `8` |

````python
a = 10
b = 3

print(a + b)   
print(a - b)   
print(a * b)   
print(a / b)   
print(a // b)  
print(a % b)   
print(a ** b)  
````

### 2.Comparison Operators
| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `==` | Equal to | `5 == 5` | `True` |
| `!=` | Not equal to | `5 != 3` | `True` |
| `>` | Greater than | `7 > 4` | `True` |
| `<` | Less than | `3 < 5` | `True` |
| `>=` | Greater or equal | `7 >= 7` | `True` |
| `<=` | Less or equal | `5 <= 8` | `True` |

````python
a = 5
b = 7
print(f"a == b -> {a == b}")
print(f"a != b -> {a != b}")
print(f"a > b -> {a > b}")
print(f"a < b -> {a < b}")
print(f"a >= b -> {a >= b}")
print(f"a <= b -> {a <= b}")
````

### 3.Logical Operators
| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `and` | Both true | `True and False` | `False` |
| `or` | Either true | `True or False` | `True` |
| `not` | Negation | `not True` | `False` |

````python
a = True
b = False
print(f"a and b -> {a and b}")
print(f"a or b -> {a or b}")
print(f"not b -> {not b}")
````

### 4. Assignment Operators
| Operator | Meaning | Example | Equivalent to |
|----------|---------|---------|----------------|
| `=` | Assign | `x = 5` | — |
| `+=` | Add and assign | `x += 3` | `x = x + 3` |
| `-=` | Subtract and assign | `x -= 2` | `x = x - 2` |
| `*=` | Multiply and assign | `x *= 4` | `x = x * 4` |
| `/=` | Divide and assign | `x /= 2` | `x = x / 2` |
| `//=` | Floor divide and assign | `x //= 2` | `x = x // 2` |
| `%=` | Modulus and assign | `x %= 3` | `x = x % 3` |
| `**=` | Exponentiate and assign | `x **= 2` | `x = x ** 2` |

````python
x = 5
print(f"x = 5  -> {x}")

x += 3   
print(f"x += 3 -> {x}")

x -= 2   
print(f"x -= 2 -> {x}")

x *= 4  
print(f"x *= 4 -> {x}")

x /= 2  
print(f"x /= 2 -> {x}")

x //= 3 
print(f"x //= 3 -> {x}")

x %= 2   
print(f"x %= 2 -> {x}")

x = 2
x **= 3  
print(f"x **= 3 -> {x}")
````

### 5. Membership Operators
| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `in` | Value exists | `"a" in "cat"` | `True` |
| `not in` | Value does not exist | `"z" not in "cat"` | `True` |

````python
fruits = ["apple", "banana", "cherry"]
word = "hello"

print(f"'apple' in fruits  -> {'apple' in fruits}")
print(f"'orange' in fruits -> {'orange' in fruits}")
print(f"'a' in word        -> {'a' in word}")
print(f"'z' not in word    -> {'z' not in word}")
````

### 6. Identity Operators
| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `is` | Same object | `x is y` | `True/False` |
| `is not` | Different objects | `x is not y` | `True/False` |


````python
a = [1, 2, 3]
b = a          
c = [1, 2, 3]  

print(f"a == b      -> {a == b}")       
print(f"a is b      -> {a is b}")       
print(f"a == c      -> {a == c}")       
print(f"a is c      -> {a is c}")       
print(f"a is not c  -> {a is not c}")   
````

### 7.round
- `round()` is a built-in Python function for rounding numbers.
- Syntax:
```python
round(number, ndigits)
```


````python
print(round(3.7))
print(round(3.2))
````

````python
pi = 3.14159
print(round(pi, 2))
print(round(pi, 3))
````

#### Special Case: Banker's Rounding
- Python uses Banker’s Rounding (round half to even).
- When the digit is exactly .5, it rounds to the nearest even number.

````python
print(round(2.5))   # 2 (not 3!)
print(round(3.5))   # 4
````
